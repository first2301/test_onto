```python
import polars.selectors as cs
import polars as pl
```

```python
pdf = pl.read_csv("./deoksan.csv")
df = pdf.clone()
```

```python
selected_numeric = df.select(cs.numeric())
selected_numeric.describe()
```

```python
skew_row = selected_numeric.select(pl.all().skew()).with_columns(statistic=pl.lit("skew"))
kurtosis_row = selected_numeric.select(pl.all().kurtosis()).with_columns(statistic=pl.lit("kurtosis"))
std_df = selected_numeric.std()
```
